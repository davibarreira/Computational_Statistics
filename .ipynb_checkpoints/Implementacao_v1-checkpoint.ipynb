Nesse notebook, vamos apresentar a implementação de modelos de volatilidade estocástica (SV). Para isso, vamos adotar a formulação de modelos de espaço-estado, onde estamos interessados em estimar a volatilidade (estado - $h_t$), que não é diretamente observável, para uma série $y_t$ realizada.  

Vamos começar com o modelo mais simples...

-----

## Modelo Canônico

<center> $ y_t = \beta \exp \left\{\frac{h_t}{2} \right\} u_t~~~$ <font color='blue'>(output - medições observadas) </font></center>

<center> $ h_t = \mu + \phi(h_{t-1} - \mu) + \sigma \eta_t  ~~~$ <font color='green'>(mudança de estado) </font></center>

<center> $h_1 \sim \mathcal{N} \left(\mu, \frac{\sigma^2}{1 - \phi^2} \right)$ </center>

Onde $h_t$ é a log-volatilidade não observada de $y_t$, e $u_t$ e $\eta_t$ são ruídos brancos gaussianos, i.e., 

$$u_t \sim \mathcal{N}(0, \sigma_u^2)~~\text{e}~~\eta_t \sim \mathcal{N}(0, \sigma_{\eta}^2)$$


Para exemplificar, vamos usar esse modelo para estimar a volatilidade do índice Ibovespa.

Tomaremos o nosso $y$ na forma de log-retorno corrigido pela média, i.e.,

$$
y_t = 100 \left\{ \left( \log IBOV_t - \log IBOV_{t-1} \right) - \frac{1}{n} \sum_{i=1}^n \left( \log IBOV_i - \log IBOV_{i-1} \right) \right\}
$$

```python
# y = log-retorno corrigido pela média
y = 100*(log_ibov - np.cumsum(log_ibov)/np.array(range(1,len(ibov))))
```

```python
# Passo 1: inicializando objetos
T = len(y)  # Nº de obs de Y
N = 1    # tamanho das amostras ("batch size")

h = np.zeros(N*T).reshape(T,N)
phi = np.zeros(N*T).reshape(T,N)
s2_n = np.zeros(N*T).reshape(T,N)
mu = np.zeros(N*T).reshape(T,N)
beta = np.zeros(N*T).reshape(T,N)

# valores iniciais
phi[0] = 0.95
s2_n[0] = 0.02
mu[0] = np.random.normal(loc=0, scale=10, size=N)  # N(0,10)
beta[0] = np.exp(mu[0]/2)
sigma_r = 5
S_sigma = 0.01 * sigma_r
```

```python
### Gibbs sampler
for t in range(1,T):
    ### Passo 2: executo até h ser aceito (acc = False)
    acc = True
    while acc:
        # amostrando ht de p(ht|h\t, y, phi, mu, s2_n)
        h_star = mu[t-1] + (phi[t-1] * ((h[t-1] - mu[t-1]) + (h[t] - mu[t-1])))/(1+phi[t-1]**2)
        v = s2_n[t-1]/(1+phi[t-1]**2)
        m = h_star + (v/2) * ((y[t-1]**2)*np.exp(-h_star)-1)
        h_prop = np.random.normal(loc=m, scale=v)  # h proposto
        
        f_star = -(1/2)*h_prop - (y[t-1]**2)/2 * np.exp(-h_prop)
        g_star = -(1/2)*h_prop - (y[t-1]**2)/2 * np.exp(-h_prop) - h_prop * np.exp(-h_star)
        
        if np.random.uniform() < f_star/g_star:
            h[t] = h_prop # aceita h[t]
            acc = False
    
    ### Passo 3:  amostrando s2_n de p(s2_n|h,y,mu,phi)
    a = (N + s2_n)
    b = S_sigma + (h[t] - mu[t-1])**2 * (1-phi[t-1]**2) + np.sum(((h[t+1] - mu[t-1])-phi[t-1]*(h[t] - mu[t-1]))**2)
    s2_n[t] = 1/np.random.gamma(shape=a/2, scale= b/2, size=N)
    
    
    ### Passo 4: amostrando phi de p(phi | h,y,mu,s2_n)
    phi_hat = np.sum((h[t+1] - mu[t-1])*(h[t] - mu[t-1]))/np.sum((h[t] - mu[t-1])**2)
    v_hat = s2_n/np.sum((h[t] - mu[t-1])**2)
    phi_prop = np.random.normal(phi_hat, v_hat, size=N)  # phi proposto
    
    # probabilidade de aceitação: e^{g_prop - g}
    priori_phi_prop = (((1+phi_prop)/2)**19) * ((1-phi_prop)/2)**0.5
    g_prop = np.log(priori_phi_prop) - ((h[0] - mu[t-1])**2*(1-phi_prop**2))/2*s2_n[t] + 1/2*np.log(1-phi_prop**2)
    
    priori_phi = (((1+phi[t-1])/2)**19) * ((1-phi[t-1])/2)**0.5
    g = np.log(priori_phi) - ((h[0] - mu[t-1])**2*(1-phi[t-1]**2))/2*s2_n[t] + 1/2*np.log(1-phi[t-1]**2)
    
    if np.random.uniform() < np.exp(g_prop - g):
        phi[t] = phi_prop
    else:
        phi[t] = phi[t-1]
        
    ### Passo 5: amostrando mu de p(mu|h,y,phi,s2_n)
    s = 1/(s2_n[t]*((N-1)*(1-phi[t])**2 + (1-phi[t]**2)))
    m = s * ((1-phi[t]**2)*h[0]/s2_n[t] + ((1-phi[t])/s2_n[t]) * np.sum(mu[t] + phi[t]*(h[t] - mu[t])+
                                       np.sqrt(s2_n[t])*np.random.standard_normal() - phi[t]*h[t]))
    print(m)
    mu[t] = np.random.normal(loc=m, scale=m)
```
